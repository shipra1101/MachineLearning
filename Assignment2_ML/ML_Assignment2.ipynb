```python
# Ques 1.1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('houses.csv')
print(df.describe())

dfs = np.split(df,[1],axis=1)
Y = dfs[0].values
X = dfs[1].as_matrix
X_1 = df['sqft_living']
print('Price '+str(Y))
print('Sqft_living feature '+str(X_1))
print('All features '+str(X))
```

Pandas Dataframe
 Pandas dataframe is a two dimensional data structure containing various rows and columns. It is a like a sql table or a spreadsheet. We can perform arithmetic operations on them.


```python
# Ques 1.2 linear regression
# normalise the data and form x, x_1, y from normalised data
sqft_living = df['sqft_living'].values
price = df['price (grands)'].values
bedrooms = df['bedrooms'].values
bathrooms = df['bathrooms'].values
sqft_lot = df['sqft_lot'].values
floors = df['floors'].values
waterfront = df['waterfront'].values
view = df['view'].values
condition = df['condition'].values
grade = df['grade'].values
sqft_above = df['sqft_above'].values
sqft_basement = df['sqft_basement'].values
age = df['age'].values
renovated_age = df['renovated_age'].values
sqft_living15 = df['sqft_living15'].values
sqft_lot15 = df['sqft_lot15'].values

sqft_living=(sqft_living-min(sqft_living))/(max(sqft_living)-min(sqft_living))
price=(price-min(price))/(max(price)-min(price))
bedrooms=(bedrooms-min(bedrooms))/(max(bedrooms)-min(bedrooms))
bathrooms=(bathrooms-min(bathrooms))/(max(bathrooms)-min(bathrooms))
sqft_lot=(sqft_lot-min(sqft_lot))/(max(sqft_lot)-min(sqft_lot))
floors=(floors-min(floors))/(max(floors)-min(floors))
waterfront=(waterfront-min(waterfront))/(max(waterfront)-min(waterfront))
view=(view-min(view))/(max(view)-min(view))
condition=(condition-min(condition))/(max(condition)-min(condition))
grade=(grade-min(grade))/(max(grade)-min(grade))
sqft_above=(sqft_above-min(sqft_above))/(max(sqft_above)-min(sqft_above))
sqft_basement=(sqft_basement-min(sqft_basement))/(max(sqft_basement)-min(sqft_basement))
age=(age-min(age))/(max(age)-min(age))
renovated_age=(renovated_age-min(renovated_age))/(max(renovated_age)-min(renovated_age))
sqft_living15=(sqft_living15-min(sqft_living15))/(max(sqft_living15)-min(sqft_living15))
sqft_lot15=(sqft_lot15-min(sqft_lot15))/(max(sqft_lot15)-min(sqft_lot15))
```

```python
# Ques 2.1 linear regression
# predict the loss curve and model for different learning rates

# code is referred from http://ozzieliu.com/2016/02/09/gradient-descent-tutorial/

def gradDescAlgo(x_1,y,theta,alpha,m,numIterations,L,hypothesis):
    xTrans = x_1.transpose()
    for i in range(0,numIterations):
        hypothesis = np.dot(x_1, theta)
        loss = hypothesis - y
        L=np.append(L,sum(loss**2))
        gradient = np.dot(xTrans, loss) / m
        theta = theta - alpha * gradient
    return theta,L,hypothesis;

x_1 = np.array([sqft_living]).T
y = price
m, n = np.shape(x_1)
numIterations= 10000
alpha = np.array([0.05,0.1,0.4])
for repeat in range(0,3):
    # L is loss array
    L=np.array([])
    # hypothesis is predicted y values
    hypothesis = np.array([])
    # theta is model parameters
    theta = np.zeros(n)
    # call gradient descent for single feature and for five different alpha
    theta,L,hypothesis = gradDescAlgo(x_1,y,theta,alpha[repeat],m,numIterations,L,hypothesis);
    # plot the graph for loss
    plt.loglog(L, label=''+str(alpha[repeat])+'X_1')

    
# predict the loss curve and model for x and y (multiple features)
x0 = np.ones(len(sqft_living));
x_all = np.array([x0,sqft_living,bedrooms,bathrooms,sqft_lot,floors,waterfront,view,condition,grade,sqft_above,sqft_basement,
                 age,renovated_age,sqft_living15,sqft_lot15]).T
m, n = np.shape(x_all)
for repeat in range(0,3):
    # L is loss array
    L=np.array([])
    # hypothesis is predicted y values
    hypothesis = np.array([])
    # theta is model parameters
    theta = np.zeros(n)
    # call gradient descent for single feature and for five different alpha
    theta,L,hypothesis = gradDescAlgo(x_all,y,theta,alpha[repeat],m,numIterations,L,hypothesis);
    # plot the graph for loss
    plt.loglog(L, label=''+str(alpha[repeat])+'X_All')

# plot learning curve of both the models in onw figure taking different learning curves in account    
plt.xlabel('Iterations'); plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Ques 1.2 MSE calculation
# choose alpha = 1 by looking at the above graph
# code is referred from http://ozzieliu.com/2016/02/09/gradient-descent-tutorial/
m, n = np.shape(x_1)
L=np.array([])
hypothesis = np.array([])
theta = np.zeros(n)
theta,L,hypothesis = gradDescAlgo(x_1,y,theta,0.1,m,numIterations,L,hypothesis);
# MSE calculation and Y prediction
Y_pred_x_1 = np.dot(x_1, theta)
loss = Y_pred_x_1 - y
mse_x_1 = (1/len(y))*sum(loss**2)
print('MSE for Model with single feature is '+str(mse_x_1))

#MSE calculation for second model with multiple features
m, n = np.shape(x_all)
L=np.array([])
hypothesis = np.array([])
theta = np.zeros(n)
theta,L,hypothesis = gradDescAlgo(x_all,y,theta,0.1,m,numIterations,L,hypothesis);
# MSE calculation and Y prediction
Y_pred_x_all = np.dot(x_all, theta)
loss = Y_pred_x_all - y
mse_x_all = (1/len(y))*sum(loss**2)
print('MSE for Model with multiple feature is '+str(mse_x_all))
```

```python
# Ques 1.3
# visualise the best obtained model for single feature
plt.scatter(x_1,y,color='r')
plt.plot(x_1,Y_pred_x_1,color='b')
plt.xlabel('Size (Sqft)'); plt.ylabel('Price ($)')
plt.show()
```

```python
# Ques 1.4
# add ridge regression with a penalty beta to it
# code is referred from http://ozzieliu.com/2016/02/09/gradient-descent-tutorial/
def gradDescRidgeAlgo(x_1,y,theta,alpha,m,numIterations,L,hypothesis,beta):
    xTrans = x_1.transpose()
    for i in range(0,numIterations):
        hypothesis = np.dot(x_1, theta)
        loss = hypothesis - y
        # add penalty beta and square of coefficents the cost
        ridge_penalty = beta*np.dot(theta.T,theta)
        # add ridge penalty in loss
        L=np.append(L,((sum(loss**2))+(ridge_penalty)))
        gradient = np.dot(xTrans, loss) / m
        theta = theta - alpha * gradient
    return theta,L,hypothesis;

x_1 = np.array([sqft_living]).T
y = price
m, n = np.shape(x_1)
numIterations= 10000
alpha = 0.1
beta = np.array([10,0.001,0.00001])
for repeat in range(0,3):
    # L is loss array
    L=np.array([])
    # hypothesis is predicted y values
    hypothesis = np.array([])
    # theta is model parameters
    theta = np.zeros(n)
    theta,L,hypothesis = gradDescRidgeAlgo(x_1,y,theta,alpha,m,numIterations,L,hypothesis,beta[repeat]);
    # plot the graph for loss
    plt.loglog(L, label=''+str(beta[repeat])+'X_1')

    
# predict the loss curve and model for x and y (multiple features)
x0 = np.ones(len(sqft_living));
x_all = np.array([x0,sqft_living,bedrooms,bathrooms,sqft_lot,floors,waterfront,view,condition,grade,sqft_above,sqft_basement,
                 age,renovated_age,sqft_living15,sqft_lot15]).T
m, n = np.shape(x_all)
for repeat in range(0,3):
    # L is loss array
    L=np.array([])
    # hypothesis is predicted y values
    hypothesis = np.array([])
    # theta is model parameters
    theta = np.zeros(n)
    theta,L,hypothesis = gradDescRidgeAlgo(x_all,y,theta,alpha,m,numIterations,L,hypothesis,beta[repeat]);
    # plot the graph for loss
    plt.loglog(L, label=''+str(beta[repeat])+'X_All')

# plot learning curve of both the models in onw figure taking different learning curves in account    
plt.xlabel('Iterations'); plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Ques 1.6
# polynomial regression of order 2
# code is referred from http://ozzieliu.com/2016/02/09/gradient-descent-tutorial/
def polynomial_regression(x_1,y,theta,alpha,m,numIterations,L,hypothesis,beta,mom):
    xTrans = x_1.transpose()
    m, n = np.shape(x_1)
    updateVectorPrev = np.zeros(n)
    for i in range(0,numIterations):
        hypothesis = np.dot(x_1, theta)
        loss = hypothesis - y
        # add penalty beta and square of coefficents the cost
        ridge_penalty = beta*np.dot(theta.T,theta)
        # add ridge penalty in loss
        L=np.append(L,((sum(loss**2))+(ridge_penalty)))
        gradient = np.dot(xTrans, loss) / m
        # add previous momentum
        momentum = mom * updateVectorPrev
        theta = theta - ((alpha * gradient)+momentum)
        # previous momentum
        updateVectorPrev = alpha * gradient
    return theta,L,hypothesis;

x0 = np.ones(len(sqft_living));
x1 = sqft_living;
x2 = (x1)**2;
#make a polynomial function
x_poly = np.array([x0,x1,x2]).T
y = price
m, n = np.shape(x_poly)
numIterations= 10000
alpha = 0.1
beta = 0.001
mom = 0.9

L=np.array([])
hypothesis = np.array([])
theta = np.zeros(n)
theta,L,hypothesis = polynomial_regression(x_poly,y,theta,alpha,m,numIterations,L,hypothesis,beta,mom);
# plot the learning curve
plt.loglog(L)
plt.xlabel('Iterations'); plt.ylabel('Loss')
plt.show()
```

```python
# MSE calculation for polynomial regression
Y_pred_x_1 = np.dot(x_poly, theta)
loss = Y_pred_x_1 - y
mse_x_1 = (1/len(y))*sum(loss**2)
print('MSE for Model with single feature is '+str(mse_x_1))
```
